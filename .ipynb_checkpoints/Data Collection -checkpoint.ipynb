```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup
```

```python
url = 'https://www.larvalabs.com/cryptopunks/details'
res = requests.get(url)
```

```python
res.content
```

```python
# create list of links to scrape 
links = []
for i in range(10_001):
    links.append(url + '/' + str(i))
links[0]
```

```python
links[2345]
```

```python
# loop through each url
# pull data from each url 
# create df
# pull data 
import time

observations = []
# start at next request
for i in range(8633,10000): 

    # request url
    one_res = requests.get(links[i], timeout=5)
    print(one_res)
    # set up BeautifulSoup 
    one_soup = BeautifulSoup(one_res.content, 'lxml')

    # table of interest 
    one_table = one_soup.find('table', {'class':'table'})
    
    # go through table and pull data of interest 
    row = one_table.find_all('tr')[1:]
    for entry in row: 
        tags = entry.find_all('td')
        data = {}
        data['url'] = links[i]
        data['type'] = tags[0].text
        data['from'] = tags[1].text
        data['to'] = tags[2].text
        data['amount'] = tags[3].text
        data['date'] = tags[4].text

        observations.append(data)
      
    print(f'we are on {i}')
    time.sleep(3)
#punks = pd.DataFrame(observations)
```

```python
new_punks = pd.DataFrame(observations)
new_punks
```
